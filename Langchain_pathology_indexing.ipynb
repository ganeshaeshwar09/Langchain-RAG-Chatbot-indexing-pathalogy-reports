```python
import openai
from openai import OpenAI
client = OpenAI()

def evaluate_response(question, rags_response, norags_response):

    evaluation_prompt = [
        {"role": "system", "content": "Evaluate the following responses based on faithfulness, relevancy, precision, and recall and just give values on scale of 0-1 for all above metrics as your a expert doctor"},
        {"role": "user", "content": f"Question: {question}"},
        {"role": "assistant", "content": f"Rags response: {rags_response}"},
        {"role": "assistant", "content": f"Norags response: {norags_response}"}
    ]
    client = OpenAI()
    completion = client.chat.completions.create(
        model="gpt-3.5-turbo",
        temperature=0,
        messages=evaluation_prompt
    )
    return completion.choices[0].message
```

```python
import time

data = []

# Assume df contains the questions and qa and llm are your model APIs

# Iterate through questions
for i in range(10):
    # Measure time and run RAGS model
    rag_t1 = time.time()
    rags_response = qa.run(df["Question"][i])
    rag_t2 = time.time() - rag_t1

    # Measure time and run LLM model
    norag_t1 = time.time()
    norags_response = llm.invoke(df["Question"][i])
    norag_t2 = time.time() - norag_t1

    # Extract fields from the response
    r = evaluate_response(df["Question"][i],rags_response,norags_response)
    fields = {}
    sections = [section.strip() for section in r.content.split('\n\n')]
    for section in sections:
        category, details = section.split(':', 1)  # Split on first colon only
        key_values = details.strip().split('\n')
        field_data = {}
        for kv in key_values:
            key, value = kv.split(': ', 1)  # Split on first ': ' to handle values with colons
            field_data[key.strip()] = float(value.strip())
        fields[category.strip()] = field_data

    # Store data for this question
    data.append({
        "Question": df["Question"][i],
        "RAGS Time": rag_t2,
        "NO RAGS Time": norag_t2,
        "Fields": fields
    })
```

```python
rows = []
for entry in data:
    row = {
        'Question': entry['Question'],
        'RAGS Time': entry['RAGS Time'],
        'NO RAGS Time': entry['NO RAGS Time'],
    }
    for key in ['Faithfulness', 'Relevancy', 'Precision', 'Recall']:
        if key in entry['Fields']:
            row[f'{key} Norags response'] = entry['Fields'][key].get('Rags response', None)
            row[f'{key} Rags response'] = entry['Fields'][key].get('Norags response', None)
        elif 'Rags response' in entry['Fields'] and 'Norags response' in entry['Fields']:
            if key in entry['Fields']['Rags response']:
                row[f'{key} Noags response'] = entry['Fields']['Rags response'][key]
            if key in entry['Fields']['Norags response']:
                row[f'{key} Rags response'] = entry['Fields']['Norags response'][key]

    rows.append(row)

# Creating the DataFrame
```

```python
import gradio as gr

def chatbot_response(query):
    # Placeholder for your actual response generation logic
    # This should return two strings:
    # 1. The response with indices (e.g., "Response [source:1]")
    # 2. The response without indices (e.g., "Response")


    response_with_index = qa.run(query)
    r = llm.invoke(query)
    response_without_index = r.content

    return response_with_index, response_without_index

# Define the Gradio interface
iface = gr.Interface(
    fn=chatbot_response,
    inputs=gr.Textbox(lines=2, placeholder="Type your query here..."),
    outputs=[
        gr.Textbox(label="Response with Index"),
        gr.Textbox(label="Response without Index")
    ],
    title="Chatbot with RAGs",
    description="This chatbot provides responses with and without indices to the source information."
)

# Launch the interface
iface.launch(debug=True)
```

```python
import streamlit as st

# Assuming you have a similar function or the necessary logic to generate responses
def chatbot_response(query):
    # Placeholder for your actual response generation logic
    # This should return two strings:
    # 1. The response with indices
    # 2. The response without indices

    response_with_index = "Response with indices [source:1]"  # Placeholder
    response_without_index = "Response without indices"  # Placeholder

    return response_with_index, response_without_index

def main():
    st.title("Chatbot with RAGs")
    st.write("This chatbot provides responses with and without indices to the source information.")

    # User input
    query = st.text_area("Type your query here...", height=150)

    # Button to generate response
    if st.button("Submit"):
        response_with_index, response_without_index = chatbot_response(query)

        # Displaying responses
        st.write("### Response with Index")
        st.write(response_with_index)

        st.write("### Response without Index")
        st.write(response_without_index)

if __name__ == "__main__":
    main()
```
